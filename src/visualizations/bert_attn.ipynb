```python
from transformers import BertTokenizer, BertModel, utils
import torch
from bertviz.neuron_view import show
from bertviz import model_view
utils.logging.set_verbosity_error()  # Suppress standard warnings
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased', output_attentions=True)
```

```python
input_text_short = "a dog"
input_text_long = "a dog is playing in the garden with a tennis ball"
```

```python
inputs = tokenizer.encode(input_text_short, return_tensors='pt')  # Tokenize input text
with torch.no_grad():
    outputs = model(inputs, output_attentions=True)  # Run model
attention = outputs[-1]  # Retrieve attention from model outputs
tokens = tokenizer.convert_ids_to_tokens(inputs[0])  # Convert input ids to token strings
model_view(attention, tokens, display_mode="light")  # Display model view
```

```python
inputs = tokenizer.encode(input_text_long, return_tensors='pt')  # Tokenize input text
with torch.no_grad():
    outputs = model(inputs, output_attentions=True)  # Run model
attention = outputs[-1]  # Retrieve attention from model outputs
tokens = tokenizer.convert_ids_to_tokens(inputs[0])  # Convert input ids to token strings
model_view(attention, tokens, display_mode="light")  # Display model view
```
