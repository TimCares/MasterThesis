```python
from validate import ZeroShotCallback
from multimodal_data2vec import KDMMData2Vec, KDMMData2VecConfig
```

```python
from datamodules import IMDBDataModule, SpeechCommandsDataModule, CIFARDataModule
```

```python
ds = SpeechCommandsDataModule(data_path='../data',
                              min_sample_size=0,
                              normalize=True,
                              pad=True,
                              batch_size=512,
                              num_workers=5,
                              shuffle=False,
                              drop_last=False)
```

```python
callback = ZeroShotCallback(n_neighbors=20, datamodules={'imdb': ds}, data_path='../data',
                            num_max_bpe_tokens=512, is_multimodal_aligned=False)
```

```python
cfg = KDMMData2VecConfig(
    depth=1,
    num_heads=12,
    mlp_ratio=0.1)
```

```python
model = KDMMData2Vec(cfg=cfg)
```

```python
from types import SimpleNamespace
d = {'model': model, 'device':'cpu'}
pl_module =  SimpleNamespace(**d)
```

```python
_, result = callback.on_validation_start(trainer=None, pl_module=pl_module)
```
