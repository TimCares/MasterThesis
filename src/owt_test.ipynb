```python
from datasets_ import OpenWebTextDataset
import os
import glob
import re
```

```python
nlp_dir_path = '../data/language'
data_path = '../data'
dataset_path = os.path.join(nlp_dir_path, 'openwebtext')
base_data_path = data_path
data_path = dataset_path
```

```python
pattern = os.path.join(nlp_dir_path, '*.tar')
files = glob.glob(pattern)

if len(files)==0:
    raise FileNotFoundError(f"No tar files found under: {dataset_path}")

print(f"Found {len(files)} tar files, inflating...")
for file in files:
    os.system(f"tar -xf {file} -C {nlp_dir_path}")
    os.remove(file)
pattern = os.path.join(dataset_path, '*.xz')
files = glob.glob(pattern)
for file in files:
    os.system(f"unxz {file}")
print("Inflated all tar files.")
```
